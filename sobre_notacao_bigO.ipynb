# Notação Big O e Análise de Algoritmos em Python


Analisar algoritmos é analisar a complexidade de diversos algoritmos e encontrar, entre eles, a solução mais eficiente para um problema bem definido. A notação Big O é uma medida estatística usada para descrever a complexidade do algoritmo.


Observação: a notação Big-O é uma das medidas usadas para a complexidade algorítmica. Outras notações incluem Big-Theta e Big-Omega. As notações Big-Omega, Big-Theta e Big-O são intuitivamente equivalentes à melhor, média e pior complexidade de tempo que um algoritmo pode atingir. Normalmente, usamos a notação Big-O como medida, em vez das outras duas, porque se pudermos garantir que um algoritmo seja executado em uma complexidade aceitável no pior caso, ele também funcionará no caso médio e no melhor caso, mas não o contrário.

Leitura: https://wiki.python.org/moin/TimeComplexity

````python
import time
import timeit
import numpy as np
import matplotlib.pyplot as plt
````

A notação Big-O significa a relação entre a entrada do algoritmo e as etapas necessárias para a execução do algoritmo, ou seja, o Big-O descreve a relação entre o tamanho da entrada (n) de um algoritmo e o limite superior do tempo de execução ou consumo de recursos, normalmente expresso em termos de funções matemáticas. É usada para analisar o desempenho assintótico de um algoritmo, ou seja, como o algoritmo se comporta à medida que o tamanho da entrada cresce para o infinito.

No entanto, a notação Big-O não está diretamente relacionada ao número absoluto de etapas necessárias para a execução de um algoritmo em uma entrada específica, mas sim à taxa de crescimento do consumo de recursos (tempo ou espaço) em relação ao tamanho da entrada. Portanto, ela fornece uma estimativa de alto nível sobre a eficiência de um algoritmo, indicando como seu desempenho se comporta à medida que a entrada aumenta.

Ela é indicada por um "O" maiúsculo seguido de um parêntese de abertura e de fechamento. Dentro do parêntese, a relação entre a entrada e as etapas executadas pelo algoritmo é apresentada usando "n".

Notação BigO: O(n)

````python
# O Big O destá função é O(n), ou seja, uma notação com complexidade linear O(n).
# Dependendo do valor de n, será definido o número de passos.
def lista1(n):
    lista = []
    for i in range(n):
        lista += [i]
    return lista

# O Big O desta função é O(1), ou seja, uma notação com complexidade constante O(c).
# Independente do valor de n, será executado 1 passo apenas.
def lista2(n):
    return list(range(n))

n = 1000
number_of_calls = 100000

# Medir o tempo de execução da função lista1
timeit_result_lista1 = timeit.timeit('lista1(n)', globals=globals(), number=number_of_calls)
print(f'Tempo de execução lista 1: {timeit_result_lista1}')
# Medir o tempo de execução da função lista2
timeit_result_lista2 = timeit.timeit('lista2(n)', globals=globals(), number=number_of_calls)
print(f'Tempo de execução lista 2: {timeit_result_lista2}')

# Comparar as funções
if timeit_result_lista1 < timeit_result_lista2:
    mais_rapida = "lista1"
else:
    mais_rapida = "lista2"

print(f"A função mais rápida foi: {mais_rapida}")
````

Big-O não está interessada em uma instância específica na qual você executa um algoritmo, mas sim em quão bem ele é dimensionado com o aumento da entrada. Essa é uma métrica muito melhor para avaliar do que o tempo concreto para uma instância concreta!.

Por exemplo, se houver uma relação linear entre a entrada e a etapa levada pelo algoritmo para concluir sua execução, a notação Big-O usada será O(n). Da mesma forma, a notação Big-O para funções quadráticas é O(n²).

O(n): em n=1, é realizada 1 etapa. Em n=10, são realizadas 10 etapas.

O(n²): em n=1, é realizada 1 etapa. Em n=10, são executadas 100 etapas.

Sendo n = 1, os dois exemplos acima teriam o mesmo desempenho. Esse é outro motivo pelo qual observar a relação entre a entrada e o número de etapas para processar essa entrada é melhor do que apenas avaliar funções com alguma entrada concreta.

````python
# Entradas de valores.
n_values = [10, 100, 1000, 10000]

# Lista para armazenar o tempo de execução de cada entrada.
timeit_results_lista1 = []
timeit_results_lista2 = []

# Iterando sobre os valores de entrada para medir o tempo de execução de cada uma e armazenando na lista.
for n in n_values:
    timeit_result_lista1 = timeit.timeit('lista1(n)', globals=globals(), number=number_of_calls)
    timeit_result_lista2 = timeit.timeit('lista2(n)', globals=globals(), number=number_of_calls)
    timeit_results_lista1.append(timeit_result_lista1)
    timeit_results_lista2.append(timeit_result_lista2)

# Criando o primeiro gráfico.
# Visualizar o tempo de execução em comparação a entrada
plt.figure(figsize=(10, 5))

plt.plot(n_values, timeit_results_lista1, label='Tempo de execução lista1')
plt.plot(n_values, timeit_results_lista2, label='Tempo de execução lista2')

plt.ylim(0, 200)
plt.xlabel('Tamanho da Entrada (n)')
plt.ylabel('Tempo de Execução (segundos)')
plt.title('Comparação de tempo de Execução')
plt.legend()
plt.grid(True)
plt.show()
````

A seguir estão algumas das funções Big-O mais comuns:

- Constante O(c)
- Linear O(n)
- Quadrático O(n²)
- Cúbico O(n³)
- Exponencial O(2ⁿ)
- Logarítmico O(log(n))
- Log Linear O(nlog(n))


**Observação:** como o Big-O modela as relações entre a entrada e as etapas, geralmente eliminamos as constantes das expressões. O(2n) é o mesmo tipo de relação que O(n) - ambos são lineares, portanto, podemos denotar ambos como O(n). As constantes não alteram a relação.

Para ter uma ideia de como um Big-O é calculado, vamos dar uma olhada em alguns exemplos de complexidade constante, linear e quadrática.

Você pode visualizar essas funções e compará-las:

### Complexidade constante - O(C)

A complexidade de um algoritmo é considerada constante se as etapas necessárias para concluir a execução de um algoritmo permanecerem constantes, independentemente do número de entradas. A complexidade constante é denotada por O(c), em que c pode ser qualquer número constante.

Vamos escrever um algoritmo simples em Python que encontre o quadrado do primeiro item da lista e, em seguida, imprima-o na tela:

````python
def constant_algo(items):
  result = items[0] * items[0]
  print(result)

constant_algo([4, 5, 6, 8])
````

No script acima, independentemente do tamanho da entrada ou do número de itens na lista de itens de entrada, o algoritmo executa apenas duas etapas:

Estaoa 1: Encontrar o quadrado do primeiro elemento

Estapa 2: Imprimir o resultado na tela

Portanto, a complexidade permanece constante, indenpendente do valor de entrada.



````python
# Lista vazia.
etapas = []

# Função com 1 etapa.
def constant(n):
  return 1

# Uma iteração que chamará a função 99 vezes e armazenará na lista o retorno 1 durante cada iteração.
for i in range(1, 100):
  etapas.append(constant(i))

# Criar um gráfico de linhas.
plt.plot(etapas, color='r', label='O(c)')

# Adicionar rótulos aos eixos e um título.
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Número de Etapas')
plt.title('Desempenho Constante - O(c)')
plt.legend()
plt.grid(True)

# Exibir o gráfico.
plt.show()
````

### Complexidade linear - O(n)


A complexidade de um algoritmo é considerada linear se as etapas necessárias para concluir a execução de um algoritmo aumentarem ou diminuírem linearmente com o número de entradas. A complexidade linear é denotada por O(n).

Neste exemplo, vamos escrever um programa simples que exibe todos os itens da lista no console:

````python
def linear_algo(items):
  for item in items:
    print(item)

linear_algo([4, 5, 6, 8])
````

A complexidade da função linear_algo() é linear no exemplo acima, pois o número de iterações do for-loop será igual ao tamanho da matriz de itens de entrada.

Por exemplo: se houver 4 itens na lista de itens, o for-loop será executado 4 vezes.


````python
etapas = []

def linear(n):
  return n

for i in range(1, 100):
  etapas.append(linear(i))

# Criar um gráfico de linhas.
plt.plot(etapas, label='O(n)')

# Adicionar rótulos aos eixos e um título.
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Número de Etapas')
plt.title('Desempenho Linear - O(n)')
plt.legend()
plt.grid(True)

plt.show()

# Atenção:
# Este código é apenas uma representação para o gerar o gráfico com desempenho linear.
````

Um aspecto importante a ser observado é que, com entradas grandes, as constantes tendem a perder valor. É por isso que geralmente removemos as constantes da notação Big-O, e uma expressão como O(2n) geralmente é abreviada para O(n). Tanto O(2n) quanto O(n) são lineares - o que importa é a relação linear, não o valor concreto. Por exemplo, vamos modificar a linear_algo():

````python
def linear_algo(items):
  for item in items:
    print(item)
  for item in items:
    print(item)

linear_algo([4, 5, 6, 8])
````

Há dois laços for que iteram sobre a lista de itens de entrada. Portanto, a complexidade do algoritmo se torna O(2n), mas no caso de infinitos itens na lista de entrada, o dobro de infinito ainda é igual a infinito. Podemos ignorar a constante 2 (já que, em última análise, ela é insignificante) e a complexidade do algoritmo continua sendo O(n).

Vamos visualizar esse novo algoritmo plotando as entradas no eixo X e o número de etapas no eixo Y:

````python
steps = []
def linear(n):
  return 2*n
for i in range(1, 100):
  steps.append(linear(i))

# Criar um gráfico de linhas.
plt.plot(steps, label='O(2n)')

# Adicionar rótulos aos eixos e um título.
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Número de Etapas')
plt.title('Desempenho Linear - O(2n)')
plt.legend()
plt.grid(True)

plt.show()
````

No script acima, você pode ver claramente que y=2n, no entanto, o resultado é linear e tem a aparência do gráfico anterior:

### Complexidade quadrática - O(n²)
A complexidade de um algoritmo é considerada quadrática quando as etapas necessárias para executar um algoritmo são uma função quadrática do número de itens na entrada. A complexidade quadrática é denotada como O(n²):

````python
def quadratic_algo(items):
  for item in items:
    for item2 in items:
      print(item, ' ', item2)

quadratic_algo([4, 5, 6, 8])
````

Temos um loop externo que percorre todos os itens da lista de entrada e, em seguida, um loop interno aninhado, que percorre novamente todos os itens da lista de entrada. O número total de etapas executadas é n*n, em que n é o número de itens na matriz de entrada.

````python
# Função que representa o algoritmo com complexidade quadrática
def quadratic_algo(items):
    steps = 0
    for item in items:
        for item2 in items:
            steps += 1
    return steps

# Tamanhos de entrada (n)
input_sizes = [10, 20, 30, 40, 50]

# Lista para armazenar o número de operações para cada tamanho de entrada
operation_counts = []

# Executar o algoritmo para cada tamanho de entrada e registrar o número de operações
for n in input_sizes:
    items = list(range(n))
    operations = quadratic_algo(items)
    operation_counts.append(operations)

# Criar um gráfico
plt.plot(input_sizes, operation_counts, label='O(n^2)')
plt.xlabel('Tamanho da Entrada (n)')
plt.ylabel('Número de Etapas')
plt.title('Complexidade Quadrática O(n^2)')
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.show()
````

Neste exemplo acima, input_sizes representa os diferentes tamanhos de entrada que você deseja testar, operation_counts registra o número de operações para cada tamanho de entrada e o gráfico mostra como o número de operações cresce de forma quadrática com o tamanho da entrada.

### Complexidade logarítmica - O(logn)

Alguns algoritmos atingem complexidade logarítmica, como o Binary Search. A pesquisa binária procura um elemento em uma matriz, verificando o meio de uma matriz e cortando a metade na qual o elemento não está. Ele faz isso novamente para a metade restante e continua as mesmas etapas até que o elemento seja encontrado. Em cada etapa, ele reduz pela metade o número de elementos da matriz.

> Isso exige que a matriz seja classificada e que façamos uma suposição sobre os dados (por exemplo, que eles estão classificados).

Quando você pode fazer suposições sobre os dados de entrada, pode tomar medidas que reduzam a complexidade de um algoritmo. A complexidade logarítmica é desejável, pois alcança um bom desempenho mesmo com uma entrada altamente escalonada.

Como descobrir a complexidade de funções complexas?
Nos exemplos anteriores, tínhamos funções bastante simples na entrada. No entanto, como calcular o Big-O de funções que chamam (várias) outras funções na entrada?

Vamos dar uma olhada:

````python
def complex_algo(items):

  for i in range(5):
    print('Python is awesome')

  for item in items:
    print(item)
  for item in items:
    print(item)

  print('Big O')
  print('Big O')
  print('Big O')

complex_algo([4, 5, 6, 8])
````

No script acima, várias tarefas estão sendo executadas:
Primeiro, uma string é impressa 5 vezes no console usando a instrução print. Em seguida, imprimimos a lista de entrada duas vezes na tela e, por fim, outra string é impressa três vezes no console.

Para descobrir a complexidade desse algoritmo, precisamos dividir o código do algoritmo em partes e tentar descobrir a complexidade das partes individuais. Anote a complexidade de cada parte.

Na primeira seção, temos:



```
for i in range(5):
    print("Python is awesome")
```

A complexidade dessa parte é O(5), pois cinco etapas constantes estão sendo executadas nesse trecho de código, independentemente da entrada.

Em seguida, temos:


```
for item in items:
    print(item)
```
Sabemos que a complexidade do trecho de código acima é O(n). Da mesma forma, a complexidade do trecho de código a seguir também é O(n):



```
for item in items:
    print(item)
```

Por fim, no trecho de código a seguir, uma string é impressa três vezes, portanto, a complexidade é O(3):


```
print("Big O")
print("Big O")
print("Big O")

```

Para encontrar a complexidade geral, basta adicionar essas complexidades individuais:
```
O(5) + O(n) + O(n) + O(3)
```

Simplificando o que foi dito acima, obtemos:
```
O(8) + O(2n) = O(8+2n)
```
Dissemos anteriormente que quando a entrada (que tem comprimento n nesse caso) se torna extremamente grande, as constantes se tornam insignificantes, ou seja, duas vezes ou metade do infinito ainda permanece infinito. Portanto, podemos ignorar as constantes. A complexidade final do algoritmo será O(n)




````python
# Tamanhos de entrada (n)
input_sizes = np.arange(1, 101)  # Por exemplo, de 1 a 100

# Lista para armazenar o número de operações (ou tempo) para cada tamanho de entrada
operation_counts = np.log(input_sizes)

# Criar um gráfico de linha
plt.plot(input_sizes, operation_counts, label='O(log n)')

# Configurações do gráfico
plt.xlabel('Tamanho da Entrada (n)')
plt.ylabel('Número de Etapas')
plt.title('Complexidade Logarítmica - (O(log n))')
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.show()
````

No código acima, input_sizes representa os tamanhos de entrada que você deseja representar. A lista operation_counts contém o número de operações (ou tempo) para cada tamanho de entrada, seguindo a função logarítmica. O gráfico mostra como o número de operações cresce de forma logarítmica à medida que o tamanho da entrada aumenta.

### Complexidade do pior caso vs melhor caso

Normalmente, quando alguém pergunta sobre a complexidade de um algoritmo, a pessoa está interessada na complexidade do pior caso (Big-O). Às vezes, ela também pode estar interessada na complexidade do melhor caso (Big-Omega).

Para entender a relação entre eles, vamos dar uma olhada em outro trecho de código:

````python
def search_algo(num, items):
  for item in items:
    if item == num:
      return True
    else:
      pass
nums = [2, 4, 6, 8, 10]

print(search_algo(2, nums))
````

No script acima, temos uma função que recebe um número e uma lista de números como entrada. Ela retorna true se o número passado for encontrado na lista de números; caso contrário, retorna None. Se você procurar por 2 na lista, ele será encontrado na primeira comparação. Esse é o melhor caso de complexidade do algoritmo, pois o item pesquisado é encontrado no primeiro índice pesquisado. A complexidade do melhor caso, nesse caso, é O(1). Por outro lado, se você pesquisar 10, ele será encontrado no último índice pesquisado. O algoritmo terá de pesquisar todos os itens da lista e, portanto, a complexidade do pior caso passa a ser O(n).

Observação: a complexidade do pior caso permanece a mesma mesmo se você tentar encontrar um elemento inexistente em uma lista - são necessárias n etapas para verificar se esse elemento não existe em uma lista. Portanto, a complexidade do pior caso continua sendo O(n).

Além da complexidade do melhor e do pior caso, também é possível calcular a complexidade média (Big-Theta) de um algoritmo, que informa "dada uma entrada aleatória, qual é a complexidade de tempo esperada do algoritmo"?

### Complexidade de espaço
Além da complexidade de tempo, em que você conta o número de etapas necessárias para concluir a execução de um algoritmo, você também pode encontrar a complexidade de espaço, que se refere à quantidade de espaço que você precisa alocar na memória durante a execução de um programa.

Dê uma olhada no exemplo a seguir:

````python
def return_squares(n):
  square_list = []
  for num in n:
    square_list.append(num * num)
  return square_list

nums = [2, 4, 6, 8, 10]
print(return_squares(nums))
````

A função return_squares() aceita uma lista de números inteiros e retorna uma lista com os quadrados correspondentes. O algoritmo precisa alocar memória para o mesmo número de itens da lista de entrada. Portanto, a complexidade de espaço do algoritmo passa a ser O(n).
